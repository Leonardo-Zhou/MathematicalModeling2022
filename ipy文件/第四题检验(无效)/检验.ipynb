```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from pprint import pprint
```

```python
initial_data = []

for i in range(5):
    initial_data.append(pd.read_excel('附件1.xlsx', i).iloc[:1000, 1:4])

```

```python
test_data = []
for i in range(12):
    test_data.append(pd.read_excel('附件2.xls',i).iloc[:1000,1:4])
```

```python
# 计算方差
std_initial = []
for j in initial_data:
    temp_outer = []
    for t in range(3):
        temp = []
        for i in range(1000):
            temp.append(np.std(j.iloc[:i,t]))
        temp_outer.append(temp)
    std_initial.append(temp_outer)
```

```python
std_target = []
for j in test_data:
    temp_outer = []
    for t in range(3):
        temp = []
        for i in range(1000):
            temp.append(np.std(j.iloc[:i,t]))
        temp_outer.append(temp)
    std_target.append(temp_outer)
```

```python
pprint(np.array(std_target))
```

```python
fig, axes = plt.subplots(3,1,figsize=(15,10))

color = ['blue','yellow','green','purple','red']

for i,ax in enumerate(axes):
    for j, std in enumerate(std_initial):
        ax.plot(std[i][50:],color=color[j])


test_i = 4
axes[0].plot(std_target[test_i-1][0][50:],color='black')
axes[1].plot(std_target[test_i-1][1][50:],color='black')
axes[2].plot(std_target[test_i-1][2][50:],color='black')


plt.show()
```
